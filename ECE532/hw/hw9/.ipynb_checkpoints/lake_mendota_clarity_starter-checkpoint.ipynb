```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('mendota_secchi_depth.txt', delimiter='\t')
x = df['day_of_year']
y = df['secchi_depth']
```

### 3.  Kernel  Regression,  Lake  Mendota  Clarity.   

The Secchi depth is  a  measure  of water clarity obtained by lowering a black and white disk off the shady side of a boatand recording the depth at which the disk is no longer visible. A dataset obtained from the University of Wisconsin’s Limnology department contains Secchi disk readings (in meters) on Lake Mendota from 2019 and 2020.  A Secchi depth of less than 2 meters is consider poor clarity, while a Secchi depth greater than 6 metersis consider very clear.  Lake Mendota can have very clear water in late spring when na-tive zoo plankton daphnia pulicariaconsume large amounts of algae and phytoplankton(for more details, see https://blog.limnology.wisc.edu/2019/06/12/whats-behind-this-extended-phase-of-crazy-clear-water-in-lake

### (i) $\lambda = 0.01$

```python
# Kernel fitting to data

lam = 0.01 #ridge regression parameter
sigma = 10
n = 30
points = 100
new_x = np.linspace(min(x), max(x), 100)
```

```python
distsq = np.zeros((n,n),dtype=float)
for i in range(0,n):
    for j in range(0,n):
        distsq[i,j]=(x[i]-x[j])**2
        
K = np.exp(-distsq/(2*sigma**2))

alpha = np.linalg.inv(K+lam*np.identity(n)) @ y
```

```python
distsq_new = np.zeros((points,n),dtype=float)
for i in range(0, points):
    for j in range(0,n):
        distsq_new[i,j] = (new_x[i] - x[j]) ** 2
K = np.exp(-distsq_new/(2*sigma**2))        
y_hat = K @ alpha
```

```python
plt.scatter(x, y, color = "black",label = "Original Data")
plt.plot(new_x, y_hat, color = "orange",label = "Kernal Fit")
plt.xlabel("Day of year")
plt.ylabel("Secchi depth")
plt.title("Original Data vs Kernal Fit (lambda = 0.01)")
plt.legend(loc = "upper right")
```

### (ii) $\lambda = 1$

```python
lam = 1

distsq = np.zeros((n,n),dtype=float)
for i in range(0,n):
    for j in range(0,n):
        distsq[i,j]=(x[i]-x[j])**2
        
K = np.exp(-distsq/(2*sigma**2))

alpha = np.linalg.inv(K+lam*np.identity(n)) @ y
```

```python
distsq_new = np.zeros((points,n),dtype=float)
for i in range(0, points):
    for j in range(0,n):
        distsq_new[i,j] = (new_x[i] - x[j]) ** 2
K = np.exp(-distsq_new/(2*sigma**2))        
y_hat = K @ alpha
```

```python
plt.scatter(x, y, color = "black",label = "Original Data")
plt.plot(new_x, y_hat, color = "orange",label = "Kernal Fit")
plt.xlabel("Day of year")
plt.ylabel("Secchi depth")
plt.title("Original Data vs Kernal Fit (lambda = 1)")
plt.legend(loc = "upper right")
```

The higher the lambda, the smoother the graph and the lesser the fit.
